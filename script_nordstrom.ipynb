```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
pwd = os.getcwd()
nordstrom_boots = pd.read_csv(pwd + '/data_nordstrom_Boots.csv')
nordstrom_boots_modified = nordstrom_boots.copy()
nordstrom_boots_modified
```

```python
nordstrom_flats = pd.read_csv(pwd + '/data_nordstrom_Flats.csv')
nordstrom_flats_modified = nordstrom_flats.copy()
nordstrom_flats_modified
```

```python
nordstrom_skirts = pd.read_csv(pwd + '/data_nordstrom_Skirts.csv')
nordstrom_skirts_modified = nordstrom_skirts.copy()
nordstrom_skirts_modified
```

```python
nordstrom_sweater = pd.read_csv(pwd + '/data_nordstrom_Sweaters.csv')
nordstrom_sweater_modified = nordstrom_sweater.copy()
nordstrom_sweater_modified
```

```python
nordstrom = pd.concat([nordstrom_boots_modified, nordstrom_flats_modified, nordstrom_skirts_modified, nordstrom_sweater_modified])
nordstrom
```

```python
nordstrom.dropna(inplace= True)
nordstrom
```

```python
category_pie = nordstrom.groupby('CATEGORY').count().loc[:, 'SUBCATEGORY']
plt.style.use('seaborn-bright')

slices = category_pie
labels = ['Clothing', 'Shoes']
slice_colors = ['#FFAE42', '#86A0B8']

plt.pie(slices, labels= labels, colors= slice_colors, wedgeprops={'edgecolor': 'black'})

plt.title('Percenge of Categories')
plt.tight_layout()
plt.show()

```

```python
subcategory_pie = nordstrom.groupby('SUBCATEGORY').count().loc[:,'CATEGORY']
plt.style.use('seaborn-bright')

slice = subcategory_pie
label = ['Boots','Flats','Skirts','Sweaters']
slice_color = ['#CD3278', '#FFE135', '#20B2AA', '#00755E']

plt.pie(slice, labels= label,colors= slice_color, wedgeprops={'edgecolor': 'black'})

plt.title('Percentage of Each Shoe Product')
plt.tight_layout()
plt.show()
```

```python
flats_filter = (nordstrom['SUBCATEGORY'] == 'Flats')

plt.style.use('seaborn-bright')

shoe_price = nordstrom[flats_filter].pivot_table(index='BRAND',values='PRICE_RETAIL',aggfunc=np.mean).sort_values('PRICE_RETAIL', ascending=0).head(10)

plt.xlabel('Price (USD)')
plt.ylabel('Brand')
plt.title('Top 10 Average Flats Price By Shoe Brand')

plt.barh(shoe_price.index, shoe_price['PRICE_RETAIL'],color='#86A0B8')
```

```python
n_num = list(nordstrom[flats_filter].loc[:,'PRICE_RETAIL'])
n = len(n_num)
n_num.sort()
if n % 2 ==0:
    median1 = n_num[n//2]
    median2 = n_num[n//2 - 1]
    median = (median1+median2)/2
else:
    median = n_num[n//2]
median
```

```python
plt.style.use('seaborn-bright')

retail_price = nordstrom[flats_filter].loc[:,'PRICE_RETAIL']

retail_price_bins = [0,100,200,300,400,500,600,700,800,900,1000]

plt.hist(retail_price, bins=retail_price_bins,color='#FFE135', edgecolor= 'black',log=True)

median_retai_price = median
color = '#E8250C'

plt.axvline(median_retai_price,color= color,label='Price Median ($110.00)', linewidth=2)
plt.legend()

plt.title('Retail Prices of Shoes')
plt.xlabel('Price')
plt.ylabel('Total Products')

plt.tight_layout()
plt.show()
```

```python
shoe_gender_pie = nordstrom.groupby(['CATEGORY', 'DEPARTMENT']).count().unstack().loc['Shoes', 'SUBCATEGORY']
plt.style.use('seaborn-bright')

gender_slice = shoe_gender_pie
gender_labels = ['Men', 'Woman']
gender_colors = ['#86A0B8', '#FD9994']

plt.pie(gender_slice, labels= gender_labels, colors= gender_colors, wedgeprops={'edgecolor': 'black'})

plt.title('Shoe Customer Gender Percentage')
plt.tight_layout
plt.show()
```

```python
plt.style.use('seaborn-bright')

x = nordstrom.loc[flats_filter, 'REVIEW_COUNT']
y = nordstrom.loc[flats_filter, 'REVIEW_RATING']
z = nordstrom.loc[flats_filter, 'PRICE_RETAIL']

plt.scatter(x, y, c=z, cmap= 'summer', edgecolors= 'black', linewidths=1, alpha=0.75)

plt.xscale('log')

cbar = plt.colorbar()
cbar.set_label('Retail Price')

plt.title('Review Rating by Review Count')
plt.xlabel('Review Count')
plt.ylabel('Review Rating')

plt.tight_layout()
```

```python
nordstrom['PRICE_TAG_DIFFERENCE'] = (nordstrom['PRICE_RETAIL'] - nordstrom['PRICE_CURRENT'])
nordstrom
```

```python
flats_color_pt = nordstrom[flats_filter].groupby('COLOR').count().loc[:,'CATEGORY'].sort_values(ascending=0).head(10)

plt.style.use('seaborn-bright')

plt.xlabel('Number of Products')
plt.ylabel('Color')
plt.title('Flats Color Count')

plt.barh(flats_color_pt.index,flats_color_pt)
```
